```python
import pandas as pd
```

```python
df = pd.read_excel("./tbgat/pattern_matching/data/ukr_populated_places_20240607.xlsx", sheet_name="Populated Places")
df2 = pd.read_csv("./tbgat/pattern_matching/data/regex_adm1.csv", sep=",")
```

```python
df
```

```python
df2
```

```python
df2_en = pd.DataFrame({"ADM_Level_EN": [None] * len(df2), "English": df2["English"]})
df2_en
```

```python
df2_ru = pd.DataFrame({"ADM_Level_RU": [None] * len(df2), "Russian": df2["Russian"]})
df2_ru
```

```python
df2_uk = pd.DataFrame({"ADM_Level_UA": [None] * len(df2), "Ukrainian": df2["Ukrainian"]})
df2_uk
```

## Small

```python
melted_en = pd.melt(df, value_name="English", var_name="ADM_Level_EN", value_vars=["ADM2_EN", "ADM1_EN"])
```

```python
melted_en = pd.concat([df2_en, melted_en], ignore_index=True)
melted_en
```

```python
melted_ru = pd.melt(df, value_name="Russian", var_name="ADM_Level_RU", value_vars=["ADM2_RU", "ADM1_RU"])
```

```python
melted_ru = pd.concat([df2_ru, melted_ru], ignore_index=True)
melted_ru
```

```python
melted_ua = pd.melt(df, value_name="Ukrainian", var_name="ADM_Level_UA", value_vars=["ADM2_UA", "ADM1_UA"])
```

```python
melted_ua = pd.concat([df2_uk, melted_ua], ignore_index=True)
melted_ua
```

```python
all = pd.concat([melted_en["English"], melted_ru["Russian"], melted_ua["Ukrainian"]], axis=1)
```

```python
all = all.drop_duplicates().reset_index(drop=True)
```

```python
all.to_csv("./tbgat/pattern_matching/generated_data/ukr_populated_places_20240607_small.csv", index=False)
```
